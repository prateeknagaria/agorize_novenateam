##Code to call microsoft cognitive API and determine sentiment of tweets

```
#importing libraries
import pandas as pd
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential
```

```
#defining access keys

key = "f7bf5ba2937b4df2a1e7f444dd26e7c6"
endpoint = "https://agorize-sentiment.cognitiveservices.azure.com/"
```

```
#defining functions

def authenticate_client():
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, 
            credential=ta_credential) 
    return text_analytics_client



def analyze_sentiment(text_sentence):
    response = client.analyze_sentiment(documents = text_sentence)[0]
    print("Document Sentiment: {}".format(response.sentiment))
    print("Overall scores: positive={0:.2f}; neutral={1:.2f}; negative={2:.2f} \n".format(
        response.confidence_scores.positive,
        response.confidence_scores.neutral,
        response.confidence_scores.negative,
    ))
    return response.sentiment, response.confidence_scores.positive, response.confidence_scores.neutral, response.confidence_scores.negative
```

```
#authenticating client
client = authenticate_client()
```

```
#reading twitter data
tweet_df = pd.read_csv("/dbfs/mnt/agorizeblob/twitter_data/twitter_extract_sg.csv")

#validating
tweet_df.tail()
```

```
text_sentiment_list = []
pos_score_list = []
neu_score_list = []
neg_score_list = []
    
for i in range(0,len(tweet_df)):
    doc_sentiment, pos_score, neu_score, neg_score = analyze_sentiment([tweet_df['Tweet Text'][i]])
    text_sentiment_list.append(doc_sentiment)
    pos_score_list.append(pos_score)
    neu_score_list.append(neu_score)
    neg_score_list.append(neg_score)
```

```
tweet_df_all_with_sentiment = tweet_df.copy()

tweet_df_all_with_sentiment['text_sentiment'] = text_sentiment_list
tweet_df_all_with_sentiment['positive_score'] = pos_score_list
tweet_df_all_with_sentiment['neutral_score'] = neu_score_list
tweet_df_all_with_sentiment['negative_score'] = neg_score_list
```

```
#saving the dataframe to datalake
tweet_df_all_with_sentiment.to_csv("/dbfs/mnt/agorizeblob/twitter_data/tweets_with_sentiment.csv",index = False)
```
